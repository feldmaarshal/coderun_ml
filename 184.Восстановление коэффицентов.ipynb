# Необходимо восстановить коэффициенты функции f(x), зная её значения на некотором наборе точек.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('data.csv', header=None)
data = data.sort_values(0)
```

```python
def f(x, a, b, c):
    return (a * np.sin(x) + b * np.log(x)) ** 2 + c * x ** 2
```

```python
(a, b, c), _ = curve_fit(f, data[0], data[1])
```

```python
plt.plot(data[0], data[1]);
```

### проверим по MSE коээфиценты

```python
mean_squared_error(data[1], f(data[0], a, b, c))
```

```python
print(f'{a:.2f}, {b:.2f}, {c:.2f}')
```
